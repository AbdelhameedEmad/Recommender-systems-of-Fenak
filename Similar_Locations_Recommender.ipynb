```python
import pandas as pd
from rake_nltk import Rake
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

#importing the dataset
df = pd.read_csv("yelp_business.csv")

#taking the attributes that we want
data = df[["business_id","name","city","categories"]]

#filling the null values with any city
data = data.fillna("Giza")

#taking a sample of the data
data = data.sample(frac=0.01, replace=True)
print(data.name.count())

#displaying the result
data.head(10)
```

```python
#setting the indicies to businesses names to call the data with
data = data.set_index('name')

# instantiating and generating the count matrix
count = CountVectorizer()
count_matrix = count.fit_transform(data['categories'])

# generating the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
# print(cosine_sim)

# creating a Series for the businesses names 
# so they are associated to an ordered numerical
# list I will use in the function to match the indexes
indices = pd.Series(data.index)

# defining the function that takes in business name
# as input and returns the top 10 recommended places
def recommendations(title, cosine_sim = cosine_sim):
    
    # initializing the empty list of recommended places
    recommended_places = []
    
    # gettin the index of the place that matches the business name
    idx = indices[indices == title].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar places
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # populating the list with the names of the best 10 matching places
    for i in top_10_indexes:
        recommended_places.append(list(data.index)[i])
        
    return recommended_places
```

```python
print(data.loc['"Cypress Street Plumbing"', 'categories'])
recommendations('"Cypress Street Plumbing"')
```
